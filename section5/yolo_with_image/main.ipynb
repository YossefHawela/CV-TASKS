```python
from PIL import Image

from ultralytics import YOLO
import matplotlib.pyplot as plt
```

```python
model = YOLO("yolo11m.pt")

path = "cat_dog.jpg"
results = model(source=path,conf=0.4)
```

```python

result_img = results[0].plot()  
plt.imshow(result_img)
plt.axis('off')
plt.show()
    
```
